```python
plt.figure(figsize=(10, 6))

ax=sns.lineplot(x='Laps', y='Position', hue='Pilote', data=position_df)
sns.set(style="darkgrid", palette="husl", font_scale=1.2)
sns.despine(left=True, bottom=True) 

ax.invert_yaxis()
plt.yticks(range(1, position_df['Position'].max() + 1))

plt.title("Position des pilotes en fonction du numÃ©ro du tour du Grand Prix d'Abu Dhabi")
plt.xlabel('NumÃ©ro du tour')
plt.ylabel('Position')
plt.legend(title='Pilote')
```
